```python
import numpy as np

def detect_outliers_z_score(data, threshold=3):
    outliers = []
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z_score = (i - mean) / std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers

scor = detect_outliers_z_score(df["StorageInGB"])

scor
```
